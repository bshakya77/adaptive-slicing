```python
import os

def count_files_in_folder(folder_path):
    # Count all files (non-directory files) in the specified folder
    file_count = sum(1 for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)))
    return file_count

# Example usage:
folder_path = '../crowdhuman_yolo/train/images'
print(f"Total files: {count_files_in_folder(folder_path)}")
```
