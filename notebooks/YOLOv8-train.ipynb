#### **Model Training for 150 epochs**

**1. Full Image Training**

```python
from ultralytics import YOLO

model = YOLO("weights/yolov8n.pt")

# Train the model
train_results = model.train(
    data="./VisDrone_bk.yaml",  # path to dataset YAML
    epochs=150,  # number of training epochs
    imgsz=640,  # training image size
    device=0,  # device to run on, i.e. device=0 or device=0,1,2,3 or device=cpu
    #freeze=10,
    lr0 = 0.001,
    lrf = 0.0001,
    dropout = 0.1,
    iou = 0.5,
    close_mosaic = 20,
    cos_lr = True
)
```

**Update the start epochs internally**

```python
import torch

# 1. Load your existing checkpoint
ckpt_path = 'runs/detect/train4/weights/last.pt'
ckpt = torch.load(ckpt_path, map_location='cpu')

# 2. Overwrite its epoch counter
ckpt['epoch'] = 150   # pretend you only trained 150 epochs so far

# 3. Save to a new checkpoint file
new_ckpt_path = 'runs/detect/train4/weights/last_epoch150.pt'
torch.save(ckpt, new_ckpt_path)
```

**2. Trained the model from checkpoint for 150 epochs on sliced datasets with freezed backbone.**

```python
#With mixed of 128 px and 64 pixels
from ultralytics import YOLO

model1 = YOLO("runs/detect/train4/weights/last.pt")

# Train the model
train_results1 = model1.train(
    data="./VisDrone.yaml",  # path to dataset YAML
    epochs=150,
    imgsz=(64, 128),
    multi_scale=True,
    freeze=10,
    lr0 = 0.001,
    lrf = 0.0001,
    weight_decay=0.001,   # regularization
    dropout=0.1,  # backbone dropout
    iou=0.5,
    close_mosaic=20,
    cos_lr=True,
    resume=False,
    device=0,
)
```

### **Analysis**

1. YOLOv11 is better than YOLOv8 when trained on full dataset and tested on full images.
2. YOLOv8 is better than YOLOv11 when trained with freezed backbone on the sliced images and tested on full images.
3. Performance degrades when the backbone is freezed compared to the full model when tested in full images in both models.
4. Performance improves when the backbone is freezed compared to the full model when tested in sliced images in both models.
5. YOLOv11n seems to be better in determing most of the categories compared to YOLOv8n.
